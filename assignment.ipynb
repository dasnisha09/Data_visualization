```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('AMONG_US.csv')
```

```python
column_team= df['Team']
```

```python
df.head()
```

```python
#i. PIE PLOT- crewmate vs imposter
plt.pie(column_team.value_counts(), labels= column_team.unique())
colors = ['lightblue', 'lightcoral']
plt.title("crewmate vs imposter")
plt.show()
```

```python
#ii. DONUT PLOT - murdered, not murdered, missing
df['Murdered'] = df['Murdered'].replace({'Yes':'Murdered', 'No': 'Not murdered','-':'Missing'})
```

```python
column=df['Murdered']
```

```python
column_count= column.value_counts()
```

```python
plt.pie(column_count,labels= column_count.index)
circle = plt.Circle((0, 0), 0.5, color='white')
plt.gca().add_artist(circle)
plt.title('DONUT CHART')
plt.show()
```

```python
df.head()
```

```python
#iii. SCATTER PLOT- sabotages fixed vs gamelength
x_data = df['Sabotages Fixed']
y_data = df['Game Length']
```

```python
plt.scatter(x_data, y_data)
plt.scatter(x_data, y_data, color='blue', marker='o', alpha=0.5)
plt.xlabel('Sabotages Fixed')
plt.ylabel('Game Length')
plt.title('Scatter Plot')
plt.show()
```

```python
#iv. HISTOGRAM- [Gamelength vs team]
import matplotlib.pyplot as plt
import numpy as np
data = np.random.normal(0, 1, 1000)  # Example data
plt.hist(data, bins=20)
plt.hist(data, bins=20, range=(-3, 3), color='blue', alpha=0.5)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.grid(True)
```

```python
#v. BIDIRECTION BAR CHART- USERID vs (WIN OR LOSS)
column1_data = df['USER ID']
column2_data = df['Outcome']
```

```python
x = range(len(column1_data))
width = 0.35

plt.bar(x, column1_data, width, label='USER ID')
plt.bar(x, column2_data, width, label='Outcome')

plt.xlabel('X-axis')
plt.ylabel('Values')
plt.title('BIDIRECTION BAR CHART')
plt.xticks(x)
plt.legend()
plt.show()
```
